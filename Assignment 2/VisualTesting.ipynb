```python
from Agent import Agent
from Environment import Environment
import random
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# This is the same code from "main.py" but slightly modified, removed the main(args)

#play game function
def playGame(env):
    while (env.gameTerminated == False):
        env.action(random.randint(1,6))
        
#create the Agent and pass into the Environment
print("Hello World")
agent = Agent(False, True, True)
env = Environment(4, 4, 0.2, False, agent, False, True)

#playing the game, pass environment
playGame(env)
```

```python
# visualizing and printing results
plt.figure(figsize=(11,11))
nx.draw_kamada_kawai(agent.graph, with_labels=True, node_size=300, font_size=20, width=2.0, node_color='lightblue')
print(agent.graph, agent.graph.edges())
```
